## Correlation Analysis

```python
total_employees = len(df)
employees_left = df['Attrition'].value_counts()['Yes']
attrition_rate = (employees_left / total_employees) * 100
avg_age = df['Age'].mean()
avg_income = df['MonthlyIncome'].mean()
avg_years = df['YearsAtCompany'].mean()

print("Total Employees:", total_employees)
print("Employees Left:", employees_left)
print("Attrition Rate:", round(attrition_rate, 2), "%")
print("Average Age:", round(avg_age, 2))
print("Average Monthly Income:", round(avg_income, 2))
print("Average Years at Company:", round(avg_years, 2))
```

# Business Insights

Insight 1

The organization has an attrition rate of 16.12%, indicating a moderate level of employee turnover.

Insight 2

Employees working overtime exhibit significantly higher attrition, suggesting that excessive workload may be a major reason for employee exits.

Insight 3

The Sales department and roles such as Sales Executive experience comparatively higher attrition than other departments and roles.

Insight 4

Employees with lower job satisfaction and poor work-life balance are more likely to leave the organization.

Insight 5

Employees receiving lower monthly income tend to exhibit higher attrition, indicating that compensation may influence employee retention.

Insight 6

Employees with shorter tenure in the organization are more likely to leave, suggesting that early-stage employee retention needs improvement.

Insight 7

Frequent business travel appears to contribute to higher attrition, potentially affecting employee well-being and job satisfaction.

Insight 8

A positive work environment and higher environment satisfaction levels are associated with lower employee attrition.

# Recommendations

Recommendation 1

The organization should monitor and reduce excessive overtime to prevent employee burnout and improve retention.

Recommendation 2

Management should focus on improving job satisfaction through employee engagement programs, recognition initiatives, and career development opportunities.

Recommendation 3

Additional attention should be given to employees in the Sales department and high-attrition job roles to identify role-specific challenges.

Recommendation 4

The company should promote a healthier work-life balance by implementing flexible work policies and wellness initiatives.

Recommendation 5

Compensation and benefits for lower-income employees should be reviewed to ensure competitiveness and improve employee retention.

Recommendation 6

A structured onboarding and mentoring program should be introduced to improve retention among newly hired employees.

Recommendation 7

Business travel policies should be reviewed to minimize employee stress associated with frequent travel.

# Conclusion

This project analyzed employee attrition patterns using exploratory data analysis techniques. The study identified several factors influencing employee attrition, including overtime, job satisfaction, work-life balance, monthly income, business travel, and employee tenure. The findings indicate that employee attrition is influenced by multiple organizational and personal factors. By implementing the recommended strategies, the organization can improve employee satisfaction, reduce turnover, and enhance workforce stability.
